```python
import pandas as pd
import os

li = []
i = 1
while os.path.exists(f"car/driving_log{i}.csv"):
    li.append(pd.read_csv(f"car/driving_log{i}.csv", names="center left right steering throttle reverse speed".split()), )
    i += 1
```

```python
df = pd.concat(li, axis=0, ignore_index=True)
```

```python
df
```

```python
for i, j in df.iterrows() :
    if any(False if os.path.exists(j[k]) else True for k in range(3)):
        
        df.drop(i, axis=0, inplace=True)
    else:
        print(True)
```

```python
df = df.reset_index(drop=True)
```

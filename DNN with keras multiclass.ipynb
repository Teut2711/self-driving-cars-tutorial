```python
import numpy as np
import tensorflow.keras as tfk
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
from sklearn import datasets, ,..
from keras.utils.np_utils import to_categorical
```

```python
n_pts = 500
centers = [[-1, 1], [-1, -1], [1, -1], [0, 0], [1, 1]]
X, y = datasets.make_blobs(n_pts, random_state=123, centers=centers, cluster_std=.4)
```

```python
plt.scatter(X[y==0, 0],X[y==0, 1])
plt.scatter(X[y==1, 0],X[y==1, 1])
plt.scatter(X[y==2, 0],X[y==2, 1])
plt.scatter(X[y==3, 0],X[y==3, 1])
plt.scatter(X[y==4, 0],X[y==4, 1])
```

```python
y_cat = to_categorical(y, 5)
```

```python
model = tfk.Sequential()
model.add(Dense(units=5, input_shape=(2,), activation='softmax'))
adam = Adam(lr=.01)
model.compile(adam, loss="categorical_crossentropy", metrics=['accuracy']) 
fitted_model = model.fit(x=X, y=y_cat, verbose=1, batch_size=50, epochs=500, shuffle='true')
```

```python
plt.plot(fitted_model.history['accuracy'],"r",fitted_model.history['loss'], "b")
plt.title('accuracy')
plt.xlabel('epoch')
plt.legend(["Accuracy", "Loss"])
plt.show()
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:, 0])-1/4, max(X[:, 0])+1/4, 50)
    y_span = np.linspace(min(X[:, 1])-1/4, max(X[:, 1])+1/4, 50)
    xx, yy = np.meshgrid(x_span, y_span)     
    xx_, yy_ = xx.ravel(), yy.ravel()
    grid = np.c_[xx_, yy_]
    pred_func = model.predict_classes(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
    
plot_decision_boundary(X, y, model)    
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.scatter(X[y==2, 0], X[y==2, 1])
plt.scatter(X[y==3, 0], X[y==3, 1])
plt.scatter(X[y==4, 0], X[y==4, 1])

prediction = model.predict(np.array([[-.5,.6]]))
print(prediction)
plt.plot(-.5, .6, marker="*",markersize=15,color="red")
```
